````python
### Preprocess the data here.
### Feel free to use as many code cells as needed.
def addRotatedImages():
    print('Generating additional data by rotating images...')
    angles = [-5, 5, -10, 10, -15, 15, -20, 20]
    new_features = []
    new_labels = []
    counts = np.bincount(y_train)
    maxcounts = max(counts)
    for i in range(len(counts)):
        multiplier = min(int(maxcounts / counts[i]) - 1, len(angles) - 1)
        if multiplier <= 1:
            continue
        mask = np.where(y_train == i)
        for j in range(multiplier):
            for feature in X_train[mask]:
                new_features.append(scipy.ndimage.rotate(feature, angles[j], reshape=False))
                new_labels.append(i)

    print("generated {} new data".format(len(new_features)))
    return new_features,new_labels
````

````python
def centerData():
    global X_train
    X_train = X_train - np.mean(X_train)
    return
````

````python
def normalizeData():
    global X_train
    X_train = X_train/255. * 0.8 + 0.1
    return
````

````python
def transform_image(img,ang_range,shear_range,trans_range):
    '''
    This function transforms images to generate new images.
    The function takes in following arguments,
    1- Image
    2- ang_range: Range of angles for rotation
    3- shear_range: Range of values to apply affine transform to
    4- trans_range: Range of values to apply translations over. 
    
    A Random uniform distribution is used to generate different parameters for transformation
    
    '''
    # Rotation

    ang_rot = np.random.uniform(ang_range)-ang_range/2
    rows,cols,ch = img.shape    
    Rot_M = cv2.getRotationMatrix2D((cols/2,rows/2),ang_rot,1)

    # Translation
    tr_x = trans_range*np.random.uniform()-trans_range/2
    tr_y = trans_range*np.random.uniform()-trans_range/2
    Trans_M = np.float32([[1,0,tr_x],[0,1,tr_y]])

    # Shear
    pts1 = np.float32([[5,5],[20,5],[5,20]])

    pt1 = 5+shear_range*np.random.uniform()-shear_range/2
    pt2 = 20+shear_range*np.random.uniform()-shear_range/2

    pts2 = np.float32([[pt1,5],[pt2,pt1],[5,pt2]])

    shear_M = cv2.getAffineTransform(pts1,pts2)
        
    img = cv2.warpAffine(img,Rot_M,(cols,rows))
    img = cv2.warpAffine(img,Trans_M,(cols,rows))
    img = cv2.warpAffine(img,shear_M,(cols,rows))
    
    return img
````

````python
def addImages():
    print('Generating additional data by transforming  images...')
    new_features = []
    new_labels = []
    angles_range = 20
    shear_range = 10
    trans_range = 5
    counts = np.bincount(y_train)
    maxcounts = max(counts)
    for i in range(len(counts)):
        if not counts[i]:
            continue
        multiplier = min (int(maxcounts / counts[i]) - 1, 10) # synthetic multiplier of ratio or max 10
        if multiplier <= 1:
            continue
        mask = np.where(y_train == i)
        for j in range(multiplier):
            for feature in X_train[mask]:
                new_features.append(transform_image(feature, angles_range, shear_range, trans_range))
                new_labels.append(i)

    print("generated {} new data".format(len(new_features)))
    return new_features,new_labels
````

````python
import tensorflow as tf

EPOCHS = 10
BATCH_SIZE = 128
````

````python
from tensorflow.contrib.layers import flatten

def TSCNet(x):    
    # Hyperparameters
    mu = 0
    sigma = 0.1
    filter_width = 5
    filter_height = 5
    colors = 3
    conv1_filters = 128
    conv2_filters = 256
    fc0_shape = 5*5*conv2_filters
    fc1_shape = 1000
    fc2_shape = 1000
    # n_classes 
    
       
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28xconv1_filters.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(filter_width, filter_height, colors, conv1_filters), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(conv1_filters))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)

    # SOLUTION: Pooling. Input = 28x28xconv1_filters. Output = 14x14xconv1_filters.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Layer 2: Convolutional. Output = 10x10x16.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(filter_width, filter_height, conv1_filters, conv2_filters), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(conv2_filters))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10xconv2_filters. Output = 5x5xconv2_filters.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Flatten. Input = 5x5xconv2_filters. Output = fc0_shape.
    fc0   = flatten(conv2)
    
    # SOLUTION: Layer 3: Fully Connected. Input = fc0_shape. Output = fc1_shape.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(fc0_shape, fc1_shape), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(fc1_shape))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)

    # SOLUTION: Layer 4: Fully Connected. Input = fc1_shape. Output = fc2_shape.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(fc1_shape, fc2_shape), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(fc2_shape))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)

    # SOLUTION: Layer 5: Fully Connected. Input = fc2_shape. Output = n_classses.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(fc2_shape, n_classes), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(n_classes))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
````

````python
#input variables
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)
````

````python
#training pipeline

rate = 0.001

logits = TSCNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
#model evaluation
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
#train the model
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    try:
        saver
    except NameError:
        saver = tf.train.Saver()
    saver.save(sess, 'tscnet')
    print("Model saved")
````

````python
#evaluate model with test data
with tf.Session() as sess:
    loader = tf.train.import_meta_graph('tscnet.meta')
    loader.restore(sess, tf.train.latest_checkpoint('./'))

    test_accuracy = evaluate(X_test, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````
